##¿Que es un dato?
####En su definición más pura, un dato es una representación simbólica (numérica, alfabética, algorítmica, espacial, etc.) de un atributo o variable cuantitativa o cualitativa.

Por sí solo, un dato no tiene un significado relevante. Es la unidad mínima de información, el bloque de construcción básico con el que construimos el conocimiento.

Tipos de datos
En el mundo de la computación y la estadística, los datos se clasifican principalmente en dos grandes grupos:

1. Datos Cualitativos (Categóricos)
Describen cualidades o características que no se pueden medir con números de forma natural.

 Nominales: Categorías sin un orden intrínseco (ej. colores, nombres de colonias en CDMX, géneros musicales).

 Ordinales: Categorías con un orden lógico (ej. nivel de satisfacción: "bajo", "medio", "alto").

2. Datos Cuantitativos (Numéricos)
Son valores que representan cantidades y permiten realizar operaciones aritméticas.

 Discretos: Valores enteros, resultantes de contar (ej. número de goles en un partido, cantidad de productos en un inventario).

 Continuos: Valores que pueden tener decimales, resultantes de medir (ej. el peso de una porción de alitas, el tiempo que tarda un repartidor).

 Datos Estructurados vs. No Estructurados
Desde una perspectiva técnica, especialmente si trabajas con bases de datos o programación:

Estructurados: Tienen un formato fijo y están organizados (como las tablas de PostgreSQL o un archivo CSV con columnas de "Precio" e "Ingreso").

No Estructurados: No tienen una forma predefinida. Ejemplos son los correos electrónicos, videos, audios o archivos PDF de reportes públicos.

Matemáticas y Estadística: Es la base para entender qué significan los números y si un resultado es válido o es simple azar. Se usan mucho las regresiones lineales y el cálculo de probabilidades.
