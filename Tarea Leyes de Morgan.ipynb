### Ley 2

# Definición del universo y los conjuntos A y B

```python
U = {1, 2, 3, 4, 5, 6}
```

```python
A = {1, 2, 3}
```

```python
B = {3, 4}
```

# Complementos

```python
A_comp = U - A
```

```python
B_comp = U - B
```

# Ley de De Morgan 1: (A ∪ B)' = A' ∩ B'

```python
union = A | B
```

```python
comp_union = U - union
```

```python
intersect_complements = A_comp & B_comp
```

```python
print("Ley de De Morgan 1:")
```

```python
print("(A ∪ B)' =", comp_union)
```

```python
print("A' ∩ B' =", intersect_complements)
```

```python
print("¿Se cumple la Ley 1?:", comp_union == intersect_complements)
```

# Ley de De Morgan 2: (A ∩ B)' = A' ∪ B

```python
intersection = A & B
```

```python
comp_intersection = U - intersection
```

```python
union_complements = A_comp | B_comp
```

```python
print("Ley de De Morgan 2:")
```

```python
print("(A ∩ B)' =", comp_intersection)
```

```python
print("A' ∪ B' =", union_complements)
```

```python
print("¿Se cumple la Ley 2?:", comp_intersection == union_complements)
```
